```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def invoice(source_path):
  !python detect.py --weights /content/drive/MyDrive/yolov9/runs/train/yolov9-c/weights/best.pt  --conf 0.3 --source {source_path}

  image_path2 = '/content/drive/MyDrive/yolov9/img_out.png'


  img1 = mpimg.imread(source_path)
  img2 = mpimg.imread(image_path2)

  fig, axs = plt.subplots(1, 2, figsize=(10, 5))

  axs[0].imshow(img1)
  axs[0].axis('off')


  axs[1].imshow(img2)
  axs[1].axis('off')

  output = pd.read_csv('output.csv')
  price = pd.read_csv('price.csv')

  merged = pd.merge(output, price, on='Products', how='left')

  merged['Amount'] = merged['Quantity'] * merged['Price']

  total = merged['Amount'].sum()
  merged = merged.append({'Products': 'Total', 'Amount': total}, ignore_index=True)

  merged.to_csv('invoice.csv', index=False)

  plt.show()
  print(merged)
```
